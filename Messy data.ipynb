# Data Validation of a super market store 

```python
#load libraries

import numpy as np
import pandas as pd
```

```python
#import data

data=pd.read_csv("Downloads/messy_data.csv")
```

```python
data.describe()
```

### Renaming  the columns name

```python
data.rename(columns={"TrnsnID":"TransactionID","Data":"Date","CusmrID":"CustomerID","ProdtID":"ProductID",
             "ProdtName":"ProductName","Qutity":"Quantity","Prize":"Price","TtlSales":"TotalSales","PaymentType":"PaymentType",
             "CusterAge":"CustomerAge"},inplace=True)
data.head()
```

### Checking the null values in data

```python
data.isnull().sum()
```

```python
# Dropping the rows of data having the null values

data.dropna(inplace=True)
```

### Checking for duplicate values

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
data.head()
```

### Editing arrays  of the data

```python
data["PaymentType"].unique()
```

### I have to merge credit with credit card and debit with debit card because they all means thesame things 

```python
data["PaymentType"]=data["PaymentType"].replace({"Debit":"Debit card","debit card":"Debit card",
                                                 "Credit":"Credit Card","paypal":"Paypal"})
```

```python
data["ProductName"].unique()
```

### Editing the product name that are wrongly spelt 

```python
data["ProductName"]=data["ProductName"].replace({"Bok N":"Book N","Boo E":"Book E","Bookk H":"Book H"})
```

### Ensuring the date are all in thesame format.

```python
data["Date"]=pd.to_datetime(data['Date'],errors='coerce')
```

```python
data.head()
```

### Upon reviewing the data, it has been observed that negative values are not valid for the Customer age column. Therefore, it is necessary to remove any rows where the Customer age is negative.

```python
data.drop(data[data["CustomerAge"] < 0].index,inplace=True)
```
